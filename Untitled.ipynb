```python
#Librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import datetime
```

```python
#Leemos el archivo meteo-nogal (las meteorológicas)
df0 = pd.DataFrame( pd.read_csv("meteo-nogal-09.csv", engine="python" ) )
df0.head()
```

```python
#Mostramos los tipos de variables de nuestros datos anteriores
df0.dtypes
```

```python
#Si nos fijamos en los tipos anteriores la mayoría dice unnamed.
#Quitaremos todas las columnas que contengan a los "unnamed".
df0=df0.drop(df0.columns[df0.columns.str.contains('unnamed:',case = False)],axis = 1)
df0.head()
```

```python
#La primera columna no nos está dando datos relevantes, sólo nos dice las unidades de medida y algunos NAN.
#Por lo que quitaremos la primer columna sin que se elimine el índice,
#haremos que el índice vuelva a contar desde cero ya que esté eliminada la columna.
df0 = df0.drop(df0.index[0]).reset_index(drop=True)
df0.head()
```

```python
#Volvemos a ver los tipos de las variables
df0.dtypes
```

```python
#Creamos una variable llamada Fecha0 apartit de Date y time.
#Que tenga al dato de Date , luego un espacio luego el dato de time. 
df0["FECHA0"] = df0["DATE"] +" "+ df0["TIME"]
df0.head()
```

```python
#Como ya tenemos FECHA0 , entonces no sirve de mucho tener a DATE y TIME por separado.
#Eliminamos DATE y eliminamos TIME.
df0.drop( ["DATE","TIME"], axis=1, inplace=True )
df0.head()
```

```python
#Queremos python que reconozca a Fecha0 como una variable de fechas.
#Creemos una variable llamada FECHA de tipo fechas a la que le asignamos el valor FECHA0. 
df0['FECHAS'] = pd.to_datetime(df0.apply(lambda x: x['FECHA0'], 1), dayfirst=True)
#Eliminemos al anterior FECHA0
df0 = df0.drop(['FECHA0'], 1)
df0.head()
```

```python
#Como ya tenemos a Fecha como una variable tipo fecha, ya podemos separarla por componentes. 
#Creamos una columna de meses
df0['MESES'] = df0['FECHAS'].dt.month
#Creamos una columna de días
df0['DIAS'] = df0['FECHAS'].dt.day
#Creamos una columna de horas
df0['HORAS'] = df0['FECHAS'].dt.hour
#Creamos una columna de minutos
df0['MINUTOS'] = df0['FECHAS'].dt.minute
df0.head()
```

```python
#Tomaremos los datos cada 30 minutos
df0 = df0[((df0['MINUTOS'] == 30.0) | (df0['MINUTOS'] == 0.0) ) & (df0['FECHAS']<'2010-01-01 00:30:00')]
#Hacemos que se recorran los índices
df0= df0.reset_index(drop=True)
df0.head()
```

```python
#Nos enfocaremos sólo en 4 de las temperaturas del subsuelo: 10cm, 20cm, 40cm, y 85cm.
#Dejaremos unicamente las columnas de años, días horas minutos
#y algunas de Tsuelo que contengan a las que mencionamos anteriormente.

df = df.filter(['2 Year_RTM  L','3 Day_RTM  L','4 Hour_Minute_RTM  L','Tsuelo_10cm','Tsuelo_20cm','Tsuelo_30cm'
,'Tsuelo_40cm','Tsuelo_55cm','Tsuelo_70cm','Tsuelo_85cm','Tsuelo_100cm'],axis=1)
df.head()
```

```python
#Necesitamos un arreglo de horas con minutos.
#convirtamos la variable 4 hour minute a una variable de tipo string
df['4 Hour_Minute_RTM  L'] = df['4 Hour_Minute_RTM  L'].astype(str)
```

```python
#Hacemos un arreglo de horas y minutos vacío
horas=[]
minutos=[]

#Iniciamos un ciclo repetitivo que vaya desde el índice 0 hasta la longitud del dataframe
for i in range (0, len(df)):
    #Si tiene 4 caracteres entonces...
    if (len(str(df['4 Hour_Minute_RTM  L'][i]))==4):
        #Si la hora es 2400 entonces...
        if (str(df['4 Hour_Minute_RTM  L'][i])[0:2]=='24'):
        #entonces le agendamos el valor de 00  a las horas
            horas.append('00')
        #entonces le agendamos el valor de 00 a los minutos
            minutos.append('00')
        #Si la hora no es 2400 entonces....
        else:
            #Le agendamos el valor de su respectivo indice de la variable las horas que ya teníamos
            horas.append(str(df['4 Hour_Minute_RTM  L'][i])[0:2])
            #Le agendamos el valor de su respectivo indice de la variable las minutes que ya teníamos
            minutos.append(str(df['4 Hour_Minute_RTM  L'][i])[2:4])
            
    #Si tiene 3 caracteres entonces.....
    elif (len(str(df['4 Hour_Minute_RTM  L'][i]))==3):
            #Le agendamos el valor de su respectivo ídice a las horas del Hour_
            horas.append(str(df['4 Hour_Minute_RTM  L'][i])[0:1])
            #Le agendamos el valor de su respectivo ídice a las horas del Hour_
            minutos.append(str(df['4 Hour_Minute_RTM  L'][i])[1:3])
            
    #Si tiene 2 caracteres entonces.....
    elif (len(str(df['4 Hour_Minute_RTM  L'][i]))==2):
            #Le agendamos el valor de 00 a horas.
            horas.append('00')
            #Le agendamos el valor de us respectivo indice de minutos al arreglo
            minutos.append(str(df['4 Hour_Minute_RTM  L'][i])[0:2])
```

```python
#Para los días al arreglo le insertamos un ciclo repetitivo con los datos de Day que teníamos
#desde el índice 0 hasta la longitud del dataframe
dias =[df['3 Day_RTM  L'][i] for i in range(0,len(df))]
```

```python
#Creamos un data frame con las horas, minutos y días del del dataframe que teníamos
f = {'dias': dias, 'horas': horas, 'minutos':minutos}
df1_fechas = pd.DataFrame(data=f)
df1_fechas.head()
```
